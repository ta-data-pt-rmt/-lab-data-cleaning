```python
import pandas as pd
```

```python
# loading the dataset
weather_df = pd.read_csv('weather-raw.csv')

#understanding the data structure

display(weather_df)
print(weather_df.columns)
```

```python
# Melting the Data: Convert the day columns into two columns where one indicates the day and the other the recorded temperature value.
melted_weather = pd.melt(weather_df, id_vars=['id', 'year', 'month', 'element'], var_name='day', value_name='temp')
melted_weather.head()
```

```python
# Pivoting the Data: Convert the melted data so that each row contains a single observation for each day, distinguishing between tmax and tmin.
pivot_weather2 = melted_weather.pivot_table(index=['id', 'year', 'month', 'day'], columns='element', values='temp', aggfunc='first')
pivot_weather = melted_weather.pivot_table(index=['id', 'year', 'month', 'day'], columns='element', values='temp', aggfunc='first').reset_index()

pivot_weather.head()
```

```python
# Check for missing values
print(pivot_weather.isnull().sum())
```

```python
pivot_weather.to_csv('tidy_weather_data.csv', index=False)
```
